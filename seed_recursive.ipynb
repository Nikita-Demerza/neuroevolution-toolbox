```python
#запустить на самоапгрейд
import sys
sys.path.append('./drive/My Drive/Colab Notebooks/seed_rl/')
import numpy as np
import pandas as pd
import tasks
import neural_tape_controller
import optimize
import pickle
import matplotlib.pyplot as plt
root_dir = './drive/My Drive/Colab Notebooks/seed_rl/'
```

```python
try:
  with open(root_dir+f'best_genoms.pkl', 'rb') as f:
      best_genoms = pickle.load(f)
  print('loaded successfully')
  genom_optimizer = best_genoms[-1]#чем оптимизируем
  genom_target = np.array(genom_optimizer)#что оптимизируем
  print('genom_target',np.shape(genom_target))
except Exception:
  nt = neural_tape_controller.nt_controller()
  genom_target = nt.nn.disassemble_genom()
  genom_target = np.array([genom_target])
  print('genom generated',np.shape(genom_target))
  with open(root_dir+f'best_genoms.pkl', 'wb') as f:
    pickle.dump(genom_target,f,protocol=pickle.HIGHEST_PROTOCOL)
```

```python
#цикл рекурсивного самоапгрейда
experinent_case = []
q_list = []
self_up_case = []
time_after_1_success_max = 15
time_after_1_success = 0
current_q = np.nan
while 1:
    print('---------new iteration----------------')
    with open(root_dir+f'best_genoms.pkl', 'rb') as f:
        best_genoms = pickle.load(f)
    print('loaded successfully')
    genom_optimizer = best_genoms[-1]#чем оптимизируем
    genom_target = np.array(genom_optimizer)#что оптимизируем
    
    
    func = tasks.full_test
    #это проигрыватель функций
    #создать контроллер
    controller = neural_tape_controller.nt_controller(tacts=1,genom=np.array(genom_optimizer))
    experiment_count = 0
    current_point = genom_target
    first_q = func(np.ravel(current_point))
    last_q = first_q
    #зарядить ленту
    controller.init_output(current_point)
    step_count = 150
    print('BEFORE')
    tasks.full_test(current_point,verbose=True,test_set=True)
    for i in range(step_count):
        #исполнить
        reward = 0
        controller.act(current_point,reward,0)
        out_tape = controller.nn.belts['out_tape'][0]
        print('i',i)
        if (out_tape[0]>0) or (len(q_list)==0):
            #трактовать это как команду на эксперимент
            current_point = np.array(out_tape)
            #ВНИМАНИЕ! Нельзя использовать 0-ое значение для каких-либо расчётов. Оно зарезервировано.
            current_point[current_point>1e3] = 1e3
            current_point[current_point<-1e3] = -1e3
            current_q = func(np.ravel(current_point),verbose=True)
            if current_q<-1e6:
              current_q=-1e6
            experiment_count+= 1
            print('current_q',current_q)
            reward = current_q - last_q
            last_q = current_q

            experinent_case.append(1)
        else:
            experinent_case.append(0)

        to_write = 0
        if len(q_list)>0:
          if current_q>np.max(q_list):
            to_write = 1
        elif len(q_list)==1:
          to_write = 1
        if to_write>0:
          #записать в файл
          time_after_1_success = time_after_1_success_max
          print('WRITTEN TO FILE')
          with open(root_dir+f'best_genoms.pkl', 'wb') as f:
              pickle.dump(np.reshape(current_point,[1,len(np.ravel(current_point))]),f,protocol=pickle.HIGHEST_PROTOCOL)

        q_list.append(current_q)
        if time_after_1_success>0:
            time_after_1_success-=1
        elif time_after_1_success==1:
            #1 - значит, пора
            time_after_1_success = 0
            #прервать, делаем самоапгрейд
            print('***********SELF-UPGRADE******************')
            self_up_case.append(1)
            break
        self_up_case.append(0)
        if len(q_list)>0:
          if i%20==0:
              plt.plot(q_list)
              plt.show()
    print('AFTER')
    tasks.full_test(current_point,verbose=True,test_set=True)
    print('experiment_count',experiment_count)
    print('last_q - first_q',last_q - first_q,last_q, first_q)
    
```
