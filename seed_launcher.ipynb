```python
from scipy import signal
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
def make_gaussian(m,s,l):
    width = int(np.ceil(s*2))
    gauss_raw = signal.gaussian(M=2*m+1+2*width, std=s)[width:]
    
    begin = 0
    end = np.min([m+width,l])
    gauss_bordered = gauss_raw[begin:end]
    if len(gauss_bordered)<=end:
        end=len(gauss_bordered)
    out = np.zeros(l-end)
    #теперь наложить gauss_bordered на out
    #out[:end] = gauss_bordered
    out = np.concatenate([gauss_bordered,out])
    return out
```

```python
def make_gaussian(m,s,l):
    #никакая не гауссиана, зато быстро
    s = np.abs(s)
    width = int(np.ceil(s))
    dx = 1./s
    gauss_raw = np.concatenate([np.arange(0,1,dx),np.arange(1,0,-dx)])

    out = np.zeros(l)
    out_start = int(m-s)
    gauss_start = 0
    out_end = int(m+s)
    gauss_end = int(s)+int(s)
    if out_start<0:
        gauss_start += -out_start
        out_start = 0
    if out_end>=l:
        gauss_end -= out_end - l
        out_end = l
        
    out[out_start:out_end] = gauss_raw[gauss_start:gauss_end]
    
    
    #теперь наложить gauss_bordered на out
    #out[:end] = gauss_bordered
    return out
```

```python
def make_gaussian(m,s,l,rng=None):
    #никакая не гауссиана, зато быстро
    #и на выходе не полноценная маска, а sparce
    s = np.abs(s)
    width = int(np.ceil(s))
    dx = 1./s
    gauss_raw = np.concatenate([np.arange(0,1,dx),np.arange(1,0,-dx)])

    out = np.zeros(l)
    out_start = int(m-s)
    gauss_start = 0
    out_end = int(m+s)
    gauss_end = int(s)+int(s)
    if out_start<0:
        gauss_start += -out_start
        out_start = 0
    if out_end>=l:
        gauss_end -= out_end - l
        out_end = l
        
    #out[out_start:out_end] = gauss_raw[gauss_start:gauss_end]
    if rng is None:
        rng = np.arange(l)
    idx = (rng>=out_start)&(rng<out_end)
    
    
    #теперь наложить gauss_bordered на out
    #out[:end] = gauss_bordered
    return gauss_raw[gauss_start:gauss_end],idx
```

```python
def make_mulydim_gaussian(m,s,l):
    #m - это векторные координаты
    #s - это вектор эллиптичноси
    #l - это измерения нашего "датакуба"
    sum_gauss = None
    for i in range(len(m)):
        gauss = make_gaussian(m[i],s[i],l[i])
        shape = [1]*i + list(np.shape(gauss))?
        gauss = np.reshape(gauss,shape)
        if not (sum_gauss is None):
            s_shape = list(np.shape(sum_gauss))
            sum_gauss = np.reshape(sum_gauss,s_shape+[1])
            sum_gauss = gauss*sum_gauss
        else:
            sum_gauss = gauss
    return sum_gauss
```

```python
m = [5,2]
s = [2,3]
l = [10,15]
out = make_mulydim_gaussian(m,s,l)
print(np.shape(out))


for i in range(10):
    plt.plot(out[:,i])
plt.show()
plt.imshow(out, cmap='hot', interpolation='nearest')
plt.show()
plt.show()
```
